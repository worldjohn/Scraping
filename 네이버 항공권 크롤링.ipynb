```python
import requests as req
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
import datetime
from datetime import timedelta
import time
```

```python
def get_item_detail(destination, url, max_cnt):
    
    browser = webdriver.Chrome()
    browser.get(url)

    time.sleep(10)
    reg = browser.page_source
    soup = bs(reg, 'html.parser')
    #print(url)

    tickets = soup.select('div.concurrent_inner__iqfJr')
    
    for i in range(len(tickets)):
        airline = tickets[i].select_one('div.airline > b.name').text
        price = tickets[i].select_one('b.item_usual__dZqAN').text
        routes = tickets[i].select('div.route_Route__2UInh')
        print(airline)
        print(price)
        print('가는편: {:10s}-> {:10s}   {:15s}'.format(routes[0].select('span.route_airport__3VT7M')[0].text, 
                      routes[0].select('span.route_airport__3VT7M')[1].text, routes[0].select_one('i.route_info__1RhUH').text))
        print('오는편: {:10s}-> {:10s}   {:15s}'.format(routes[1].select('span.route_airport__3VT7M')[1].text, 
                      routes[1].select('span.route_airport__3VT7M')[1].text, routes[1].select_one('i.route_info__1RhUH').text))
        print('--------'*5)    
        
        if i >= max_cnt - 1:
            break
    
    browser.quit()
```

```python
from_date = str(datetime.date.today() + timedelta(days=1)).replace('-', '')
to_date   = str(datetime.date.today() + timedelta(days=30)).replace('-', '')

destinations = {'뉴욕': 'JFK', '싱가포르': 'SIN', '시드니': 'SYD'}

max_cnt = 10
is_direct = True
for key, value in destinations.items():
    direct = '&isDirect=true' if is_direct else ''
    url = f'https://flight.naver.com/flights/international/ICN-{value}-{from_date}/{value}-ICN-{to_date}?adult=1{direct}&fareType=Y'

    print('*****'*10)
    print('목적지:', key)
    print('*****'*10)
    get_item_detail(key, url, max_cnt)
    
print('크롤링이 끝났습니다.')
```
