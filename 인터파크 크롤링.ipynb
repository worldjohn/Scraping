```python
import requests as req
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
```

```python
def get_item_detail(tab, url, max_cnt):
    
    res = req.get(url)
    soup = bs(res.text, 'html.parser')
    
    items = soup.select('dl.Tview')
    for i in range(len(items)):
        print(items[i].select_one('dd > span.txt1').text)
        print(items[i].select_one('dd > span.txt2').text)
        print(items[i].select_one('dd > span.txt3').text)
        image = items[i].select_one('dt > a > span > img')
        if image != None:
            img_url = image['src']

            with urlopen(img_url) as f:
                with open('./images/ticket/' + tab + str(i+1)+'.jpg','wb') as h:
                    data = f.read()
                    h.write(data)
                    
        print('--------'*5)
        if i >= 4:
            break
```

```python
tabs = {'뮤지컬': 'Musical', '콘서트': 'Concert', '연극': 'Drama', '클래식': 'Classic'}

max_cnt = 5
for key, value in tabs.items():
    url = f'https://ticket.interpark.com/{value}Index.asp'
    
    print('*****'*10)
    print('*** {}'.format(key))
    print('*****'*10)
    get_item_detail(key, url, max_cnt)
    print()
    
    #if key == '뮤지컬': break

print('크롤링이 끝났습니다.')
```
